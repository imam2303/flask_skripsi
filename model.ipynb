```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv("dataset/pmks_dataset1.csv")
```

```python
df.head()
```

```python
df.drop('nama', axis=1, inplace=True)
df.drop('alamat', axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['jenis_pmks'] = label_encoder.fit_transform(df['jenis_pmks'])
df['hubungan_dlm_keluarga'] = label_encoder.fit_transform(df['hubungan_dlm_keluarga'])
df['jml_tanggungan_kepala_keluarga'] = label_encoder.fit_transform(df['jml_tanggungan_kepala_keluarga'])
df['pendapatan_keluarga'] = label_encoder.fit_transform(df['pendapatan_keluarga'])
df['status_rumah'] = label_encoder.fit_transform(df['status_rumah'])
df['pekerjaan'] = label_encoder.fit_transform(df['pekerjaan'])
```

```python
df
```

```python
X = df.drop("layak_tidak", axis=1)
y = df["layak_tidak"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nbc = GaussianNB()
nbc.fit(X_train, y_train)
```
